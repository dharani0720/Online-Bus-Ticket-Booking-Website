```python
class Student:
    def __init__(self, name, course):
        self.name = name
        self.course = course

    def display(self):
        print("Student Name:", self.name)
        print("Course:", self.course)
        print()

student1 = Student("Dharani", "Diploma in Computer Engineering")
student2 = Student("Karthik", "Diploma in Mechanical Engineering")

student1.display()
student2.display()
```
